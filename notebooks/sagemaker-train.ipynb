# Script para treinamento em super máquina com stagemaker

```python
import sagemaker
import boto3

from sagemaker.pytorch import PyTorch
```

```python
sess = sagemaker.Session()
inputs = {
    "train": "s3://hirise-dtm-fill/datasets/train/",
    "validation": "s3://hirise-dtm-fill/datasets/validation/"
}
```

```python
estimator = PyTorch(
    entry_point="marsfill/entrypoint/train.py",
    source_dir="../",
    role=sagemaker.get_execution_role(),
    instance_type="ml.g5.12xlarge",
    distribution={"pytorchddp": {"enabled": True}},
    instance_count=1,
    image_uri="109762776171.dkr.ecr.us-east-1.amazonaws.com/marsfill-training-env:latest",
    hyperparameters={},
    use_spot_instances=True,
    max_run=43200,
    max_wait=43200 + 3600
)
```

```python
estimator.fit(inputs)
```
